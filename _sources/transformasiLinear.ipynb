### **Transformasi Linier**
- Transformasi $=$ fungsi $=$ pemetaan (mapping)                 
Misalkan V dan W adalah ruang vektor. Transformasi yang memetakan ruang vektor $\vee$ ke ruang vektor $\mathrm{W}$ ditulis sebagai
$
\mathrm{T}: \mathrm{V} \rightarrow \mathrm{W}
$
 
V adalah daerah asal (domain) transformasi T dan W adalah daerah hasil transformasi (kodomain) fungsi. 

Jika $\mathrm{V}=\mathrm{W}$, maka $\mathrm{T}$ dinamakan operator pada V.
- Jika $\mathbf{v} \in \mathrm{V}$ dan $\mathbf{w} \in \mathrm{W}$, maka
$
\mathbf{w}=\mathrm{T}(\mathbf{v})
$
      
Misalkan $\mathrm{T}: \mathrm{R}^3 \rightarrow \mathrm{R}^3$ didefinisikan sebagai berikut:
$$
T\left(\left[\begin{array}{l}
x_1 \\
x_2 \\
x_3
\end{array}\right]\right)=\left[\begin{array}{c}
x_1+2 x_2+x_3 \\
x_1+5 x_2 \\
x_3
\end{array}\right]
$$

Tentukan bayangan vektor $\mathbf{v}=(3,2,0)$.
Jawaban:
$$
T\left(\left[\begin{array}{l}
3 \\
2 \\
0
\end{array}\right]\right)=\left[\begin{array}{c}
3+2(2)+0 \\
3+5(2) \\
0
\end{array}\right]=\left[\begin{array}{c}
7 \\
13 \\
0
\end{array}\right]
$$

Jadi, bayangan vektor $(3,2,0)$ adalah $(7,13,0)$.

 Diberikan fungsi $T: R^2 \rightarrow R^2$ yang dalam hal ini $T(x, y)=(2 x, y)$, maka akan ditunjukkan bahwa T adalah transformasi linier.
Misalkan $\mathbf{u}$ dan $\mathbf{v}$ adalah dua buah vektor di $\mathrm{R}^2, \mathbf{u}=\left(\mathbf{u}_1, \mathrm{u}_2\right)$ dan $\mathbf{v}=\left(\mathrm{v}_1, \mathbf{v}_2\right)$                              
(1)                                            
$$
\begin{aligned}
T(\mathbf{u}+\mathbf{v}) & =\mathrm{T}\left(u_1+v_1, u_2+v_2\right)=\left(2\left(u_1+v_1\right), u_2+v_2\right)=\left(2 u_1+2 v_1, u_2+v_2\right) \\
& =\left[\begin{array}{c}
2 u_1+2 v_1 \\
u_2+v_2
\end{array}\right]=\left[\begin{array}{c}
2 u_1 \\
u_2
\end{array}\right]+\left[\begin{array}{c}
2 v_1 \\
v_2
\end{array}\right]=T(u)+T(\mathbf{v})
\end{aligned}
$$                                                                          
                                                        
                                                                                                       
                                                                                                        
(2)                                                    
$$
\begin{aligned}
& \mathrm{T}(k \mathbf{u})=\mathrm{T}\left(k u_1, k u_2\right)=\left(2 k u_1, k u_2\right) \\
&=\left[\begin{array}{c}
2 k u_1 \\
k u_2
\end{array}\right]=k\left[\begin{array}{c}
2 u_1 \\
u_2
\end{array}\right]=k \mathrm{~T}(\mathbf{u})
\end{aligned}
$$
                                                     
                                                               
                                                               
Karena $T(\mathbf{u}+\mathbf{v})=\mathrm{T}(\mathbf{u})+\mathrm{T}(\mathbf{v})$ dan $\mathrm{T}(k \mathbf{u})=k T(\mathbf{u})$, maka $T$ adalah transformasi linier

### Refleksi
Refleksi  merujuk pada transformasi yang memantulkan vektor-vekstor melalui suatu garis atau bidang. Dalam konteks dua dimensi, refleksi dilakukan terhadap garis tertentu (misalnya, sumbu x atau sumbu y), sedangkan dalam tiga dimensi, refleksi dilakukan terhadap bidang tertentu (misalnya, bidang xy, yz, atau zx).

### Contoh Matrik Refleksi dalam 2D

1. **Refleksi terhadap sumbu x:**
   Matrik refleksi terhadap sumbu x akan membalikkan tanda komponen y dari vektor, sedangkan komponen x tetap. Matrik refleksi ini adalah:
   $$
   R_x = \begin{pmatrix}
   1 & 0 \\
   0 & -1
   \end{pmatrix}
   $$

2. **Refleksi terhadap sumbu y:**
   Matrik refleksi terhadap sumbu y akan membalikkan tanda komponen x dari vektor, sedangkan komponen y tetap. Matrik refleksi ini adalah:
   $$
   R_y = \begin{pmatrix}
   -1 & 0 \\
   0 & 1
   \end{pmatrix}
   $$

3. **Refleksi terhadap garis y = x:**
   Matrik refleksi terhadap garis y = x akan menukar komponen x dan y dari vektor. Matrik refleksi ini adalah:
   $$
   R_{y=x} = \begin{pmatrix}
   0 & 1 \\
   1 & 0
   \end{pmatrix}
   $$

4. **Refleksi terhadap garis y = -x:**
   Matrik refleksi terhadap garis y = -x akan menukar komponen x dan y dari vektor, serta membalikkan tanda keduanya. Matrik refleksi ini adalah:
   $$
   R_{y=-x} = \begin{pmatrix}
   0 & -1 \\
   -1 & 0
   \end{pmatrix}
   $$

### Contoh Matrik Refleksi dalam 3D

1. **Refleksi terhadap bidang xy:**
   Matrik refleksi terhadap bidang xy akan membalikkan tanda komponen z dari vektor, sedangkan komponen x dan y tetap. Matrik refleksi ini adalah:
   $$
   R_{xy} = \begin{pmatrix}
   1 & 0 & 0 \\
   0 & 1 & 0 \\
   0 & 0 & -1
   \end{pmatrix}
   $$

2. **Refleksi terhadap bidang yz:**
   Matrik refleksi terhadap bidang yz akan membalikkan tanda komponen x dari vektor, sedangkan komponen y dan z tetap. Matrik refleksi ini adalah:
   $$
   R_{yz} = \begin{pmatrix}
   -1 & 0 & 0 \\
   0 & 1 & 0 \\
   0 & 0 & 1
   \end{pmatrix}
   $$

3. **Refleksi terhadap bidang zx:**
   Matrik refleksi terhadap bidang zx akan membalikkan tanda komponen y dari vektor, sedangkan komponen z dan x tetap. Matrik refleksi ini adalah:
   $$
   R_{zx} = \begin{pmatrix}
   1 & 0 & 0 \\
   0 & -1 & 0 \\
   0 & 0 & 1
   \end{pmatrix}
   $$

### Contoh Aplikasi Refleksi
Misalkan kita memiliki vektor $( \mathbf{v} = \begin{pmatrix} 3 \\ 4 \end{pmatrix} )$. Jika kita ingin merefleksikan vektor ini terhadap sumbu y, kita gunakan matriks $( R_y )$:

$$
R_y = \begin{pmatrix}
-1 & 0 \\
0 & 1
\end{pmatrix}
$$

Maka hasil refleksi vektor $( \mathbf{v} )$ terhadap sumbu y adalah:

$$
R_y \mathbf{v} = \begin{pmatrix}
-1 & 0 \\
0 & 1
\end{pmatrix}
\begin{pmatrix}
3 \\
4
\end{pmatrix}
=
\begin{pmatrix}
-3 \\
4
\end{pmatrix}
$$

Jadi, vektor hasil refleksi adalah $$ \begin{pmatrix} -3 \\ 4 \end{pmatrix} $$
